```python
import pandas as pd

# Datos del escenario
data = {
    "País": ["A", "B", "C", "D", "E"],
    "Tamaño Mercado (M$)": [120, 85, 180, 65, 140],
    "Crecimiento Proyectado (%)": [15, 22, 8, 18, 12],
    "Intensidad Competitiva": ["Alta", "Media", "Baja", "Baja", "Media"],
    "Costo Entrada (M$)": [8.5, 5.2, 12.0, 4.0, 7.8],
    "Riesgo Regulatorio": ["Bajo", "Medio", "Alto", "Bajo", "Medio"]
}

df = pd.DataFrame(data)
print(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))
ax = sns.scatterplot(
    data=df,
    x="Costo Entrada (M$)",
    y="Tamaño Mercado (M$)",
    hue="Intensidad Competitiva",
    style="Riesgo Regulatorio",
    s=300
)

# Etiquetas de cada punto
for i, row in df.iterrows():
    ax.text(row["Costo Entrada (M$)"]+0.2, 
            row["Tamaño Mercado (M$)"],
            row["País"],
            fontsize=12,
            weight="bold")

plt.title("Relación entre Tamaño de Mercado y Costo de Entrada")
plt.xlabel("Costo de Entrada (M$)")
plt.ylabel("Tamaño de Mercado (M$)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()

```

```python
# Ventas proyectadas
ventas = {
    "A": [100, 125, 160],
    "B": [100, 140, 190],
    "C": [100, 115, 135],
    "D": [100, 130, 170],
    "E": [100, 120, 145]
}

ventas_df = pd.DataFrame(ventas, index=["Año 1", "Año 2", "Año 3"])

# Gráfico
ventas_df.plot(marker="o", figsize=(8,6))
plt.title("Proyecciones de Ventas a 3 Años (Índice Relativo)")
plt.ylabel("Ventas (índice)")
plt.show()
```

```python
plt.figure(figsize=(8,6))

# Ordenamos el DataFrame por crecimiento proyectado
df_sorted = df.sort_values("Crecimiento Proyectado (%)", ascending=False)

sns.barplot(
    data=df_sorted,
    y="País",
    x="Crecimiento Proyectado (%)",
    palette="viridis"
)

plt.title("Crecimiento Proyectado por País (ordenado)")
plt.xlabel("Crecimiento Proyectado (%)")
plt.ylabel("País")
plt.show()

```

```python
# Convertimos riesgos e intensidades en valores numéricos para heatmap
mapa = df[["País", "Intensidad Competitiva", "Riesgo Regulatorio"]].copy()

intensidad_map = {"Baja": 1, "Media": 2, "Alta": 3}
riesgo_map = {"Bajo": 1, "Medio": 2, "Alto": 3}

mapa["Intensidad Competitiva"] = mapa["Intensidad Competitiva"].map(intensidad_map)
mapa["Riesgo Regulatorio"] = mapa["Riesgo Regulatorio"].map(riesgo_map)

mapa.set_index("País", inplace=True)

plt.figure(figsize=(6,4))
sns.heatmap(mapa, annot=True, cmap="YlOrRd", cbar=True)
plt.title("Mapa de Intensidad Competitiva y Riesgo Regulatorio")
plt.show()
```
